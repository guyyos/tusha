```python
import numpy as np

# Set the random seed for reproducibility
np.random.seed(42)

# Number of groups and observations per group
G = 10
N = 50

# Number of individual-level predictors
p = 3

# Generate group-level predictors
Z = np.random.normal(loc=0, scale=1, size=(G, p))

# Generate individual-level predictors
X = np.random.normal(loc=0, scale=1, size=(G * N, p))

# Generate group-level parameters
beta_0 = np.random.normal(loc=0, scale=1)
beta_z = np.random.normal(loc=0, scale=1, size=p)

# Generate individual-level parameters
beta = np.random.normal(loc=0, scale=1, size=(G * N, p))

# Generate the response variable
y = beta_0 + np.sum(beta_z * Z, axis=1) + np.sum(beta * X, axis=1)

# Add noise to the response variable
epsilon = np.random.normal(loc=0, scale=0.1, size=G * N)
y += epsilon
```
